```python
import pandas as pd
import matplotlib.pyplot as plt

reference = pd.read_csv('results/data.csv');
real = pd.read_csv('results/trajectory.csv');

import matplotlib as mpl
import matplotlib.patches as patches
from matplotlib.collections import PatchCollection
import math
from colour import Color
import numpy as np

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111)
ax.set_aspect("equal")

length = 0.31 / cell_size
width = 0.2 / cell_size

def create_state_rectangle(x, y, theta, speed):
    rect = plt.Rectangle((-0.5*length, -0.5*width), length, width)
    trans = mpl.transforms.Affine2D().rotate_deg((theta / math.pi) * 180) + mpl.transforms.Affine2D().translate(x, y)
    rect.set_transform(trans)
    
    return rect

def create_rectangles(data):
    rectangles = []
    for index, state in data.iterrows():
        rect = create_state_rectangle(state["x"] / cell_size, state["y"] / cell_size, state["heading_angle"], state["speed"])
        rectangles.append(rect)
    return rectangles

collection = PatchCollection(create_rectangles(real), edgecolor="black", linewidth=1.0)
ax.add_collection(collection)

plt.grid()
plt.imshow(grid,interpolation='nearest',cmap='gray', alpha=0.1)
plt.plot(reference["x"] / cell_size, reference["y"] / cell_size, alpha=0.3, c="green")
plt.plot(real["x"] / cell_size, real["y"] / cell_size, alpha=0.3, c="blue")

plt.scatter(waypoints[0] / cell_size, waypoints[1] / cell_size, c="red")

plt.show()
```

```python
import math

def circle_from(row):
    return {
        "x": row["x"],
        "y": row["y"],
        "radius": row["radius"]
    }

def distance(point_a, point_b):
    dx = point_a["x"] - point_b["x"]
    dy = point_a["y"] - point_b["y"]
    return math.sqrt(dx ** 2 + dy ** 2)

def are_directly_visible(circle_a, circle_b):
    dist = distance(circle_a, circle_b)
    dx = cell_size * (circle_b["x"] - circle_a["x"]) / dist
    dy = cell_size * (circle_b["y"] - circle_a["y"]) / dist

    point = circle_from(circle_a);
    while distance(point, circle_b) >= cell_size:
        point["x"] = point["x"] + dx
        point["y"] = point["y"] + dy
        
        x = int(point["x"] / cell_size)
        y = int(point["y"] / cell_size)
                
        if not grid[y][x]:
            return False
    
    return True
```

```python
raw_circles = [circle_from(row) for i, row in circles.iterrows()]
raw_circles.append(raw_circles[0]) # we assume it's a circuit

filtered_circles = [raw_circles[0]]
last_circle = raw_circles[0]
prev_circle = raw_circles[0]

for circle in raw_circles[1:]:
    if not are_directly_visible(last_circle, circle):
        filtered_circles.append(prev_circle)
        last_circle = prev_circle
    prev_circle = circle

filtered_circles
```

```python
free = []
obstacle = []

def mark_steps_between(circle_a, circle_b):
    dist = distance(circle_a, circle_b)
    dx = cell_size * (circle_b["x"] - circle_a["x"]) / dist
    dy = cell_size * (circle_b["y"] - circle_a["y"]) / dist

    point = circle_from(circle_a);
    while distance(point, circle_b) >= cell_size:
        point["x"] = point["x"] + dx
        point["y"] = point["y"] + dy
        
        x = int(point["x"] / cell_size)
        y = int(point["y"] / cell_size)

        if grid[y][x]:
            free.append([x, y])
        else:
            obstacle.append([x, y])

last_circle = raw_circles[0]

for circle in filtered_circles[1:]:
    mark_steps_between(last_circle, circle)
    last_circle = circle
```
