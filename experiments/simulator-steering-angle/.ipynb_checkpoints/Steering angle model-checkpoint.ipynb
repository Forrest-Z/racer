# Data description

In the experiment, we set the throttle and the steering angle to a fixed value for 10 seconds and we let the vehicle move on an obstacle-free plane. After 10 seconds, we slightly changed the steering angle or the throttle. We would systematically try all the different steering angles and throttles.


Every 0.5s we marked the position of the center of vehicle's rear axle. Using the last three positions, we calculated the radius of the circle the vehicle was circumscribing. Every 10 seconds, when the steering input changed, we calculated the average of the 20 measurements and we printed one line to the CSV file.

```python
import math

# wheelbase
L = 0.32

def steering_angle(radius):
  return math.atan(L / radius)

def steering_angles(radii):
  angles = []
  for radius in radii:
      angles.append(steering_angle(radius))
  return angles

# maximum steering angle - the lowest speed and the maximum steering wheel input
min_radius = min(csv["radius"])
theta = steering_angle(min_radius)
math.degrees(theta)
```

```python
throttles = list(set(csv["throttle"].to_list()))
throttles.sort()
throttles
```

```python
def analyze_steering_angles_at(throttle):
    data = csv[csv["throttle"] == throttle]
    left = data[data["steering"] > 0]
    right = data[data["steering"] < 0]
    plt.title("throttle: {}".format(throttle))
    plt.plot(left["steering"], left["radius"], "b-")
    plt.plot(right["steering"], right["radius"], "r-")
    plt.show()

for throttle in throttles:
    analyze_steering_angles_at(throttle)
```

```python

import numpy as np

def fit_radius(throttle):
    data = csv[csv["throttle"] == throttle]
    left = data[(data["steering"] > 0.4) & (data["steering"] < 0.85)]

    x = left["steering"]
    # y = left["radius"]
    y = np.array([steering_angle(r) for r in left["radius"]])

    A = np.vstack([x, np.ones(len(x))]).T
    m, c = np.linalg.lstsq(A, y, rcond=None)[0]

    p = np.poly1d(np.polyfit(x, y, 2))

    plt.title("throttle = {}".format(throttle))
    plt.plot(x, y, 'o', label='Original data', markersize=5)
    plt.plot(x, m*x + c, 'r', label='Fitted line: {0:.3f}*x + {1:.3f}'.format(m, c))
    plt.plot(x, p(x), 'g', label='Fitted quadratic curve')
    plt.legend()
    plt.show()
    
    return m, c
    
ms = []
cs = []
for throttle in throttles:
    m, c = fit_radius(throttle)
    ms.append(m)
    cs.append(c)
```

```python
x_ = np.array(throttles)
y_ = np.array(ms)
above = np.where(x_ >= 0.4)
x = x_[above]
y = y_[above]

A = np.vstack([x, np.ones(len(x))]).T
a, b = np.linalg.lstsq(A, y, rcond=None)[0]

below = np.where(x_ < 0.4)
u = x_[below]
v = y_[below]

A = np.vstack([u, np.ones(len(u))]).T
m, n = np.linalg.lstsq(A, v, rcond=None)[0]

plt.plot(throttles, ms, 'r.', label='m')
plt.plot(x, a*x + b, 'g-', label='fitted: m = {0:0.3f}*t + {1:0.3f}'.format(a, b))
plt.plot(u, m*u + n, 'b-', label='fitted: m = {0:0.3f}*t + {1:0.3f}'.format(m, n))
plt.legend()
plt.show()
```

```python
def steering_angle_estimate(throttle, steering):
    if throttle < 0.4:
        m = 1.074 * throttle + 0.406
        c = -0.582 * throttle + 0.095
        return m * steering + c
    else:
        m = -0.086 * throttle + 0.229
        c = -0.036 * throttle + 0.026
        return m * steering + c
```

```python
def test_estimate(throttle):
    data = csv[csv["throttle"] == throttle]
    left = data[data["steering"] > 0.0]

    estimated = [steering_angle_estimate(throttle, s) for s in left["steering"]]
    measured = np.array([steering_angle(r) for r in left["radius"]])
                 
    plt.title("throttle = {}".format(throttle))
    plt.plot(left["steering"], measured, 'b.', label='Measured steering angles')
    plt.plot(left["steering"], estimated, 'r-', label='Estimated steering angles')
    plt.legend()
    plt.show()
    
for throttle in throttles:
    test_estimate(throttle)
```
