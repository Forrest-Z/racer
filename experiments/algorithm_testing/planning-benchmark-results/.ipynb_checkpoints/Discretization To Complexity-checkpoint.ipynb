```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = ["porto.csv", "race_track.csv", "simple.csv", "u.csv", "tornado.csv", "zurich.csv"]

for name in names:
    csv = pd.read_csv(name)
    successful = csv[csv["found solution"] == "yes"]
    successful

    sehs = successful[successful["algorithm"] == "sehs"]
    hybrid = successful[successful["algorithm"] == "hybrid_astar"]

    plt.title(name)
    plt.plot(sehs["time to finish"], sehs["average computation time in ms"], "b.", label="SEHS")
    plt.plot(hybrid["time to finish"], hybrid["average computation time in ms"], "r.", label="Hybrid A*")
    plt.plot([min(successful["time to finish"]), max(successful["time to finish"])], [min(successful["average computation time in ms"]), max(successful["average computation time in ms"])], ls="--", c=".3")
    plt.xlabel("time to finish [s]")
    plt.ylabel("computation time [ms]")
    plt.show()
```

```python
import re

max_rpm = 17500

def complexity(row):
    a = float(row[1]["number of actions"])
    d = row[1]["discretization"]
    x = re.search("rpm-([\d.]+)-theta-([\d.]+)", d)
    rpm = float(x.group(1))
    theta = float(x.group(2))
    return a * (max_rpm / rpm) * (360.0 / theta)
    
def complexities(rows):
    return [complexity(r) for r in rows.iterrows()]
```

```python
actions = 88

def analyze(name):
    csv = pd.read_csv(name)
    successful = csv[csv["found solution"] == "yes"]
    successful

    sehs = successful[successful["algorithm"] == "sehs"]
    sehs_higlight = sehs[(sehs["number of actions"] == actions) & (sehs["discretization"].str.endswith("rpm-875-theta-20"))]
    hybrid = successful[successful["algorithm"] == "hybrid_astar"]
    hybrid_highlight = hybrid[(hybrid["number of actions"] == actions) & (hybrid["discretization"].str.endswith("rpm-875-theta-20"))]

    plt.title(name)
    plt.plot(complexities(hybrid), hybrid["time to finish"], ".", label="Hybrid A*", color='0.65')
    plt.plot(complexities(sehs), sehs["time to finish"], ".", label="SEHS", color="0.65")
    plt.plot(complexities(hybrid_highlight), hybrid_highlight["time to finish"], "b.", label="Chosen params Hybrid A*")
    plt.plot(complexities(sehs_higlight), sehs_higlight["time to finish"], "r.", label="SEHS with chose params")
    plt.xlabel("complexity")
    plt.ylabel("time to finish [ms]")
    plt.legend()
    plt.show()

    plt.title(name)    
    plt.plot(complexities(hybrid), hybrid["average computation time in ms"], ".", label="Hybrid A*", color="0.65")
    plt.plot(complexities(sehs), sehs["average computation time in ms"], ".", label="SEHS", color="0.65")
    plt.plot(complexities(hybrid_highlight), hybrid_highlight["average computation time in ms"], "b.", label="Hybrid A*")
    plt.plot(complexities(sehs_higlight), sehs_higlight["average computation time in ms"], "r.", label="SEHS with chose params")
    plt.ylabel("computation time [ms]")
    plt.xlabel("complexity")
    plt.legend()
    plt.show()

for name in names:
    analyze(name)
```
