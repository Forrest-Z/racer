```python
data = pd.read_table("./2019-10-26-02/10x_max-1000ms_l-3.csv", sep=',', skipinitialspace=True)
len(data)
```

```python
combined_data = []

for i in range(0, int(len(data) / 7)):
    s = i * 7
    if data.iloc[s]['found solution'] == 'no':
        continue

    total = 0
    for j in range(i * 7, (i + 1) * 7):
        total += data.iloc[j]['average computation time in ms']
    combined_data.append({'sample': data.iloc[i], 'mean_computation_time': total / 7.0})

combined_data
```
